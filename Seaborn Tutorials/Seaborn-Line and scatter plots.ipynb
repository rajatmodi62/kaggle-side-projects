Hello. This is my beginning on kaggle.Many of us face problems with understanfing the nature of data. This post covers the basis of Seaborn Plotting library. Seaborn is used as a supplement to matlabplotlib for higher level of understanding.Lets start by importing the seaborn library 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

The first concept is off scatterplot. When we have two numeric features, then we can visualize them as clouds of information in two dimensional plane.(X-Y axis)

```python
tips=sns.load_dataset("tips")
tips
```

As you can see the total_bill,tips and size are the numberic features. Lets try to plot two out of them 

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

As an additional information , a colour dimension can be added signifying the placement of a third variable 

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

we can see that the red coloured dots and blue coloured dots represent different smoking categories, and is plotted on total_bills vs tips. Suppose we want to change the shape of each dot.


```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips);
```

If the hue variable is categorical (like smokers yes or no), the colour which gets plotted takes a form of random colours. (Qualitative Palette which is more pleasing to look at). But if the hue variable is numeric(size), then palette takes sequential values as we trasnittion from one value to another)

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);

```

So we saw two variables can be plotted with each other, thid can be added via hue variable , nature of thepalette changes acc to wether hue is categorical or numeric feature. Next, out of the points plotted some might be more or less important.
Lets plot those 


```python
sns.relplot(x="total_bill", y="tip",hue="size", size="size", data=tips);

```

so we see that the colour and size of each point changes as per the importance (from low to higher value of the size variable)

**Handling uncertainity in the data**

Sometimes in multiple numneric features, correspoinding to same feature we might have multiple values which reperesent a degree of uncertainity. That time the aggregation comes to solution. Lets fire up the fmri dataset


```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

For each timepoint, many values of signals occur,lets make a simeple line plot

```python
g=sns.relplot(x="timepoint",y="signal",kind="line",data=fmri)
```

The greyed out region shows us some region of uncertainity, which is the default behavioru of seaborn.You can see that at lower values we have lower uncertainity in this dataset.By deafult, a mean is figured out for each variable and a region of 95% confidence interval taken into play.

However for large datasets, predicting this uncertainity requires caclulation of mean and standard deviation for each variable which is timeconsuming.So lets us disable it using:

```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',ci=None)
```

Uncertainity goes away :-). But there is a problem. Corresponding to ach point on x axis there were mutiple values of y. But here the plot is continuos. It means that the averaging is still going on behind the scenes. We need to stop that. Therefore we use,estimator attribute.


```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',estimator=None,ci=None)
```

Instead of mean as a confidecne interval , we could have used the standard deviation.

```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',ci='sd')
```

Let us now extrapolate the conceps of hue to the scatterplots. We can make line plots for multiple variables

```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line')
```

```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue="event",ci='sd')
```

```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue="event",style="region",ci='sd')
```

So we have plotted two different line plots for the dependence of signal and timepoints variables to events. In each event, we are analyzing the variation of parietal and frontal regions. 

The point to note for the line plots is that the x axis and y axis should be numeric features, hue and style can be categorical features as well. The number of unique values in hue should be low, otherwise the colour palette will run out while plotting. 

By adding the marker flag .we are now able to visualize what the unique points at which dataset was plotted

```python
g=sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue="event",style="region",ci='sd',markers=True)
```

***Understanding the concept of faceting ****

Plotting multiple variables in the scatter plot makes the figure complex.We can split the figures using Faceting.
Variables vary across a row and column , and for each of such values we plot the dependence of two numberical values wrt to each other.For eg consider the mri dataset 

```python
fmri
```

```python
g=sns.relplot(x='timepoint',y='signal',
             kind='line',hue='event',
            style='event', col='subject', data=fmri)
```

A lots of plots came in single line , lets wrap them up

```python
g=sns.relplot(x='timepoint',y='signal',
             kind='line',hue='event',
            style='event', col='subject', data=fmri,
             col_wrap=5)
```

As you can see that the plot is for different subjects. But the values are still not visible. We need some mechanism to zoom in on the plot.

```python
g=sns.relplot(x='timepoint',y='signal',
             kind='line',hue='event',
            style='event', col='subject', data=fmri,
            height=3, aspect=.75, linewidth=2.5,
             col_wrap=5)
```

Summary:
- Scatter plots give discrete distribution of data.Line plots give continous ones after estimating mean or standard deviation for points correspoding the the same values. There is a degree of extrapolation involved in line plots which gives them their contonous nature
- Hue can split the scatter/line plots for multiple variables, and style markes the data distribution in diffeent shaped dots
- Faceting can be used to plot multiple features
